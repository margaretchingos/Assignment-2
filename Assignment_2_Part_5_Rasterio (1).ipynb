##### Task 1: Import data from google drive. Read the file as a Rasterio dataset

#### Task 2: What is the CRS of the dataset?

```python
print('the coordinate reference system of the raster is')
print(raster.crs) #this prints the coordinate reference system of the raster
```

#### Task 3:  Describe the raster dataset regarding the raster extent (bounds), the reference system, and how many bands are in this dataset.

```python
bounds = (raster.bounds) #this defines 'bounds' as the boundaries of the raster
print('the raster extent is:', bounds) #this prints a sentence that describes the bounds of the raster using 'bounds' as defined above
crs = (raster.crs) #repeat for crs
print('the coordinate reference system of the raster is', crs)
bandcount = (raster.count) #repeat for bandcount
print('the raster has', bandcount, 'band')
```

#### Task 4: Create a plot/map of the raster dataset.

```python
from rasterio.plot import show #this imports the show function to make maps from rasterio
show(raster) #this uses the show function to make a plot of 'raster'. Becuase there is only one band in the raster I do not have to specify which band to map
```

#### Task 5:  Create Histograms from the raster

```python
from rasterio.plot import show_hist #this imports the function of rasterio plot to make histograms
show_hist(raster, bins=40, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled', title="Histogram") #this uses the show_hist function to make a histogram of 'raster'.
#this histogram has 40 bins (meaning it splits up the data into 40 groups). According to this, there are no outliers.
#I will make a histogram with more bins to see if outliers come up.
```

```python
show_hist(raster, bins=300, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled', title="Histogram")
#this histogram has 300 bins. It still does not show outliers, so I do not need to mask any outliers to correctly represent my data
```
