```python
#import libraries 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#import data
df = pd.read_csv('/arcgis/home/Assignment 2/portland_park_trees.csv')
#preview first 5 rows of data
df.head()
```

#### Graph 1: Create a scatterplot for just trees in the Ulmus genus with DBH mapped to the x-axis and tree height mapped to the y-axis 

```python
ulmus = df.query('Genus=="Ulmus"') #subset the genus ulmus

x = ulmus["DBH"] #set ulmus DBH to x
y = ulmus["TreeHeight"] #set ulmus Tree Height to y

plt.scatter(x,y) #create a scatter plot of x and y
plt.title("Ulmus DBH vs Tree Height") #change the title
plt.xlabel("DBH") #change the x axis label
plt.ylabel("Tree Height") #change the y axis label
```

#### Graph 2: Create a scatterplot for just trees in the Ulmus genus with DBH mapped to the x-axis, tree height mapped to the y-axis, and tree species mapped to hue

```python
#I will use seaborn for this graph
#Import seaborn:
import seaborn as sns

#make a plot from ulmus where x=DBH, y=Tree Height and the species are all different colors
sns.relplot(x="DBH", y="TreeHeight", hue="Species", data=ulmus)
plt.title("Ulmus DBH vs Tree Height") #give the graph a title
```

#### Graph 3: Create a boxplot of DBH for just the Ulmus genus differentiated by species

```python
ulmus = df.query('Genus=="Ulmus"') #subset the genus ulmus

sns.boxplot(data=ulmus, x="Species", y="DBH") #make a boxplot of DBH by species 
plt.title('Ulmus DBH by Species') #give the boxplot a title
```

#### Graph 4: Combine graphs 1 and 3

```python
ulmus = df.query('Genus=="Ulmus"') #query the dataframe to include just the ulmus genus
x = ulmus["DBH"] #set ulmus DBH to x
y = ulmus["TreeHeight"] #set ulmus Tree Height to y


fig, (ax1, ax2) = plt.subplots(1,2) #creates a figure with 2 subplots
ax1.scatter(x, y) #ax1 (subplot 1) will be a scatterplot of x,y
sns.boxplot(data=ulmus, x="Species", y="DBH", ax=ax2) #ax2 (subplot 2) will be a boxplot of ulmus DBH by species. Becuase this is a seaborn graph, the ax is defined at the end, as opposed to the matplotlib graph in ax1. 
```
